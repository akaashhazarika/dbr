```python
def get_path_between_nodes(k, start, end):

  fib = [0]* 48

  fib[0] = 0
  fib[1] = 1
  for i in range(2, 48):
    fib[i] = fib[i-1] + fib[i-2]


  def get_path_preorder(k, target_value):
    if k == 0: return []

    starting = 1 # Root is at the very beginning
    path = []

    while starting != target_value:
      path.append(starting)
      left_size = fib[k+1] - 1

      # If target is within the left subtree's range
      if target_value <= starting + left_size:
        starting = starting + 1            # Left root is immediately next
        k = k - 1
      else:
        starting = starting + left_size + 1  # Right root skips the left subtree
        k = k - 2

    return path + [target_value]

  def get_path_inorder(k, target_value):
    if k==0:
      return []
    starting = (fib[k+1]-1)+1
    path = []
    while starting!=target_value:
      path.append(starting)
      if target_value > starting:
        starting = starting + fib[k-1]
        k = k-2
      else:
        starting = (starting-1) - (fib[k-1]-1)
        k = k-1

    return path + [target_value]


  def get_path_postorder(k, target_value):
    if k==0:
      return []
    left_elements = fib[k+1]-1
    right_elements = fib[k] - 1
    starting = left_elements + right_elements +1
    path = []
    while starting!=target_value:
      left_elements = fib[k+1]-1
      right_elements = fib[k] -1
      path.append(starting)
      if target_value < starting - right_elements:
        starting = starting-right_elements-1
        k = k-1
      else:
        starting = starting-1
        k = k-2
    return path + [target_value]



  start_node_path =  get_path_postorder(k, start)
  end_node_path = get_path_postorder(k, end)

  print(start_node_path)
  print(end_node_path)
  i = 0
  while i<len(start_node_path) and i<len(end_node_path) and start_node_path[i] == end_node_path[i]:
    i = i+1
  return list(reversed(start_node_path[i-1:]))+end_node_path[i:]

get_path_between_nodes(4,1,5)
```
